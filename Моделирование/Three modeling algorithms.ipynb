```python
import math
import random
import matplotlib.pyplot as plt
import numpy as np

    
def inverseFunctionMethod():
    alpha = random.random()
    a = math.sqrt(math.exp(1))
    b = 8 + 33 * a
    d = a / b
    f = 8 / b
    g = f + 4 * d
    h = g + 4 * d
    i = 1 - d
    j = (8 - alpha * b) / a
    if alpha < f:
        return (-1 * math.sqrt(-2 * math.log(alpha / (8 * d))))
    elif alpha < g:
        return (-1 * math.sqrt(1 + j / 4))
    elif alpha < h:
        return (math.sqrt(-1 - j / 4))
    elif alpha < i:
        return (math.sqrt(- j / 8))
    else:
        return (2 / math.sqrt(33 + j))
    

def discreteDecompositionMethod():
    alpha1, alpha2 = random.random(), random.random()
    a = math.exp(1 / 2)
    b = 8 + 33 * a
    q1 = 8 / b
    q2 = (8 + 4 * a) / b
    q3 = 8 * (1 + a) / b
    q4 = 1 - a / b
    if alpha1 < q1:
        return (-1 * math.sqrt(1 - 2 * math.log(alpha2)))
    elif alpha1 < q2:
        return (-1 * math.sqrt(1 - alpha2))
    elif alpha1 < q3:
        return (math.sqrt(alpha2))
    elif alpha1 < q4:
        return(math.sqrt(3 * alpha2 + 1))
    else:
        return(2 / math.sqrt(1 - alpha2))
    

        
def selectionMethod():
    f = 1 - math.atan(1 / 2) / math.pi
    while True:
        alpha1, alpha2 = random.random(), random.random()
        a = 1 / math.tan(math.pi * (1 - alpha1))
        b = a ** 2 + 1
        c = a * b
        if not((alpha1 < 1/4 and -1 * c * math.exp(-1 / 2 * a ** 2) < 20 * alpha2) or \
        (1 / 4 < alpha1 < 1 / 2 and -1 * c < 20 * alpha2) or \
        (1 / 2 < alpha1 < 3 / 4 and c < 20 * alpha2) or (3 / 4 < alpha1 < f and 2 * c < 20 * alpha2) or \
        (alpha1 > f and b < 20 * alpha2 * a ** 3)):
            #print(alpha1)
            return a        


sizeOfSample = 10000
x1 = np.arange(-4, -1, 0.01)
x2 = np.arange(-1, 0, 0.01)
x3 = np.arange(0, 1, 0.01)
x4 = np.arange(1, 2, 0.01)
x5 = np.arange(2, 4, 0.01)
c = 8 * math.exp(1 / 2) /  (8 + 33 * math.exp(1 / 2))


#Метод обратных функций.
fig, ax = plt.subplots(1, 1)
ax.grid()
ax.set_xlim([-4, 4])
ax.set_ylim([0, 0.9])
ax.plot(x1, c * (-1) * x1 * np.exp(-1 * x1 ** 2 / 2), x2, c * (-1) * x2, x3, c * x3, x4, c * 2 * x4, x5, c * x5 ** (-3), color="red")

sampleIFM = [inverseFunctionMethod() for i in range(sizeOfSample)]
ax.hist(sampleIFM, 1000, density=True, color="gray")
plt.title("Метод обратных функций")
fig.savefig("./inverseFunctionMethod.png")

# Метод дискретной декомпозиции.
fig, ax = plt.subplots(1, 1)
ax.grid()
ax.set_xlim([-4, 4])
ax.set_ylim([0, 0.9])
ax.plot(x1, c * (-1) * x1 * np.exp(-1 * x1 ** 2 / 2), x2, c * (-1) * x2, x3, c * x3, x4, c * 2 * x4, x5, c * x5 ** (-3), color="red")

sampleDDM = [discreteDecompositionMethod() for i in range(sizeOfSample)]
ax.hist(sampleDDM, 1000, density=True, color="gray")
plt.title("Метод дискретной декомпозиции")
fig.savefig("./discreteDecompositionMethod.png")


# Метод отбора.
fig, ax = plt.subplots(1, 1)
ax.grid()
ax.set_xlim([-4, 4])
ax.set_ylim([0, 0.9])
ax.plot(x1, c * (-1) * x1 * np.exp(-1 * x1 ** 2 / 2), x2, c * (-1) * x2, x3, c * x3, x4, c * 2 * x4, x5, c * x5 ** (-3), color="red")

sampleSM = [selectionMethod() for i in range(sizeOfSample)]
ax.hist(sampleSM, 1000, density=True, color="gray")
plt.title("Метод отбора")
fig.savefig("./selectionMethod.png")

plt.show()

```

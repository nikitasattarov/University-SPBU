```python
import math
import random
import matplotlib.pyplot as plt
import numpy as np

    
def inverseFunctionMethod():
    alpha = random.random()
    c = math.exp(-1)
    b = 6 + c
    a = c / b
    d = (c + 3 / 2) / b
    g = (c + 11 / 2) / b
    if alpha < a:
        return (math.log(alpha * b))
    elif alpha < d:
        return (1 - math.sqrt(2) * math.sqrt(2 + c - alpha * b))
    elif alpha < g:
        return (-1 + math.sqrt(2) * math.sqrt(-1 - c + alpha * b))
    else: 
        return(1 / (6 + c - alpha * b))
    
    

def discreteDecompositionMethod():
    alpha1, alpha2 = random.random(), random.random()
    a = math.exp(-1)
    b = 1 / (a + 6)
    q1 = a * b
    q2 = (3 / 2 + a) * b
    q3 = q2 + 4 * b
    if alpha1 < q1:
        return (math.log(alpha2) - 1)
    elif alpha1 < q2:
        return (1 - math.sqrt(4 - 3 * alpha2))
    elif alpha1 < q3:
        return (-1 + math.sqrt(1 + 8 * alpha2))
    else:
        return(2 / (1 - alpha2))
    

        
def selectionMethod():
    f = 1 - math.atan(1 / 2) / math.pi
    while True:
        alpha1, alpha2 = random.random(), random.random()
        a = 1 / math.tan(math.pi * (1 - alpha1))
        b = a ** 2 + 1
        c = 1 + a
        d = 1 - a
        if not((alpha1 < 1 / 4 and b * math.exp(a) < 15 * alpha2) or \
        (1 / 4 < alpha1 < 1 / 2 and b * d < 15 * alpha2) or \
        (1 / 2 < alpha1 < f and b * c < 15 * alpha2) or (alpha1 > f and b < 15 * alpha2 * a ** 2)):
            #print(alpha1)
            return a        

        
sizeOfSample = 10000
x1 = np.arange(-4, -1, 0.01)
x2 = np.arange(-1, 0, 0.01)
x3 = np.arange(0, 2, 0.01)
x4 = np.arange(2, 4, 0.01)
c = 1 / (math.exp(-1) + 6)


#Метод обратных функций.
fig, ax = plt.subplots(1, 1)
ax.grid()
ax.set_xlim([-4, 4])
ax.set_ylim([0, 0.9])
ax.plot(x1, c * np.exp(x1), x2, c * (-x2 + 1), x3, c * (x3 + 1), x4, c * x4 ** (-2), color="red")

sampleIFM = [inverseFunctionMethod() for i in range(sizeOfSample)]
ax.hist(sampleIFM, bins = 10000, density=True, color="gray")
plt.title("Метод обратных функций")
fig.savefig("./inverseFunctionMethod.png")

# Метод дискретной декомпозиции.
fig, ax = plt.subplots(1, 1)
ax.grid()
ax.set_xlim([-4, 4])
ax.set_ylim([0, 0.9])
ax.plot(x1, c * np.exp(x1), x2, c * (-x2 + 1), x3, c * (x3 + 1), x4, c * x4 ** (-2), color="red")

sampleDDM = [discreteDecompositionMethod() for i in range(sizeOfSample)]
ax.hist(sampleDDM, bins = 10000, density=True, color="gray")
plt.title("Метод дискретной декомпозиции")
fig.savefig("./discreteDecompositionMethod.png")


# Метод отбора.
fig, ax = plt.subplots(1, 1)
ax.grid()
ax.set_xlim([-4, 4])
ax.set_ylim([0, 0.9])
ax.plot(x1, c * np.exp(x1), x2, c * (-x2 + 1), x3, c * (x3 + 1), x4, c * x4 ** (-2), color="red")

sampleSM = [selectionMethod() for i in range(sizeOfSample)]
ax.hist(sampleSM, bins = 10000, density=True, color="gray")
plt.title("Метод отбора")
fig.savefig("./selectionMethod.png")

plt.show()
```
